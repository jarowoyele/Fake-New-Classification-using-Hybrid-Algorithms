```python
data_true = pd.read_csv("/content/drive/My Drive/Dad Damilola/Fake News Dataset/fakenews/True.csv")
data_fake = pd.read_csv("/content/drive/My Drive/Dad Damilola/Fake News Dataset/fakenews/Fake.csv")
```

```python

data_true['Target'] = ['True']*len(data_true)
data_fake['Target'] = ['Fake']*len(data_fake)
```

```python

dataset = data_true.append(data_fake).sample(frac=1).reset_index().drop(columns=['index'])
```

```python
dataset.head(10)
```

```python
dataset["Target"].value_counts()
```

```python
dataset['label'] = pd.get_dummies(dataset.Target)['Fake']
```

```python
dataset.head(10)
```

```python
def removal(text):
    text = re.sub('<[^>]*>', '', text)
    emoticons = re.findall('(?::|;|=)(?:-)?(?:\(|D|P)',text.lower())
    text = re.sub('[\W]+', ' ', text.lower())
    text += ' '.join(emoticons).replace('-', '')
    return text
```

```python
dataset["text"] = dataset["text"].apply(removal)
```

```python
X = dataset["text"]
y = dataset["label"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
X_transformedtrain = vectorization.fit_transform(X_train)
X_transformedtest = vectorization.transform(X_test)
```

```python
X_transformedtrain.shape, X_transformedtest.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier 
```

```python
log_classifier = LogisticRegression()
log_classifier.fit(X_transformedtrain, y_train) 
```

```python
decision_classifier = DecisionTreeClassifier()
decision_classifier.fit(X_transformedtrain, y_train) 
```

```python
decision_classifier.score(X_transformedtest,y_test)
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
voting_ensemle = VotingClassifier(estimators = [("log_classifier",LogisticRegression()),
                                                ("decision_classifier",DecisionTreeClassifier())], voting = "hard")
```

```python
voting_ensemle.fit(X_transformedtrain, y_train)
```

```python
voting_ensemle.score(X_transformedtest, y_test)
```
